```python
from gigachat import GigaChat
import os
api_key = os.getenv("GIGACHAT_API_KEY")

# Укажите ключ авторизации, полученный в личном кабинете, в интерфейсе проекта GigaChat API
llm = GigaChat(credentials=api_key, verify_ssl_certs=False)
response = llm.chat("Какие факторы влияют на стоимость страховки на дом?")
print(response.choices[0].message.content)
```
